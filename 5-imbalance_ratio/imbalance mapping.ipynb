```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_file_name = './data/Imbalance Ratio.csv'
imbalance = pd.read_csv(input_file_name, sep = ",")
# drop the first unamed column 
imbalance = imbalance.drop(imbalance.columns[0], axis = 1)
imbalance = imbalance.rename(columns={"station number": "Station"})
imbalance[:10]
```

```python
imbalance['ratio'] = 10*imbalance[['diff']].div(imbalance[['bikes out', 'bikes in']].min(axis=1), axis=0)
imbalance['class'] = np.where(imbalance['diff']>=0, 1, 0)
imbalance
```

```python


import json
import pandas as pd
with open('../map-inventory.json', 'r') as f:
    map_station = json.load(f)
map_station = map_station['features']
map_station_data = []
# w/o station name
# for each in map_station:
#     map_station_data.append([int(each['properties']['station']['terminal']),each['geometry']['coordinates'][0],each['geometry']['coordinates'][1]]) 
# map_station_data= pd.DataFrame(map_station_data, columns = ['Station', 'longitude', 'latitude'])

# w/ station name
for each in map_station:
    map_station_data.append([int(each['properties']['station']['terminal']),each['properties']['station']['name'],each['geometry']['coordinates'][0],each['geometry']['coordinates'][1]]) 
map_station_data= pd.DataFrame(map_station_data, columns = ['Station', 'name','longitude', 'latitude'])

# map_station_data
```

```python
imbalance_map = pd.merge(imbalance, map_station_data, on='Station')
imbalance_map[:20]
```

```python
output_file_name_map = './data/imbalance ratio map.csv'
imbalance_map.to_csv(output_file_name_map)
```
